```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df1 = pd.read_csv('1.병원정보서비스 2022.10..csv', encoding = 'CP949')
df2 = pd.read_csv('2.약국정보서비스 2022.10..csv', encoding = 'CP949')
df3 = pd.read_csv('4.의료기관별상세정보서비스_02_세부정보_2022.10..csv', encoding = 'CP949')
df4 = pd.read_csv('5.의료기관별상세정보서비스_03_진료과목정보_2022.10..csv', encoding = 'CP949')
```

```python
# 주소에서 동만 추출해서 '동'에 저장
# '읍면동'에 있고 '동'에 없는 값은 합쳐줌
def find_dong(x):
    left = x.find('(')
    right = x.find('동') + 1
    return x[left + 1:right]

for i in range(len(df1)):
    if (df1.loc[i, '읍면동'] != 'NaN') & (df1.loc[i, '동'] == ''):
        df1.loc[i, '동'] = df1.loc[i, '읍면동']
```

```python
df1.isnull().sum()
```

```python
df2['동'] = df2['주소'].apply(lambda x : find_dong(x))

from tqdm import tqdm

for i in tqdm(range(len(df2))):
    if (df2.loc[i, '읍면동'] != 'NaN') & (df2.loc[i, '동'] == ''):
        df2.loc[i, '동'] = df2.loc[i, '읍면동']
```

```python
# '오시는 길' 컬럼 생성
df3 = pd.read_csv('4.의료기관별상세정보서비스_02_세부정보_2022.10..csv', encoding = 'CP949')

df3['위치_공공건물(장소)명'].fillna('', inplace = True)
df3['위치_방향'].fillna('', inplace = True)
df3['위치_거리'].fillna('', inplace = True)

df3['오시는 길'] = df3['위치_공공건물(장소)명'] + ' ' + df3['위치_방향'] + ' ' + df3['위치_거리']
df3['오시는 길']

df3_new = df3.drop(['위치_공공건물(장소)명', '위치_방향', '위치_거리'], axis = 1)
df3_new.head()
```

```python
# '휴진 안내' 컬럼 생성 (일요일 휴진안내 + 공휴일 휴진 안내)
df3_new['일요일 휴진안내'].fillna('없음', inplace = True)
df3_new['공휴일 휴진안내'].fillna('없음', inplace = True)
df3_new['휴진 안내'] = '[휴진 안내]' + '\n' + '일요일 : ' + df3_new['일요일 휴진안내'] + '\n공휴일 : ' + df3_new['공휴일 휴진안내']
df3_new['휴진 안내']
```

```python
# 이상치 변경
df3_new.loc[4025, '응급실 주간운영여부'] = 'Y'
```

```python
# 응급실 운영 여부 N/Y 로 맞춰줌
df3_new['응급실 주간운영여부'].fillna('N', inplace = True)
df3_new['응급실 야간운영여부'].fillna('N', inplace = True)

df3_new['응급실 주간운영여부'] = df3_new['응급실 주간운영여부'].apply(lambda x : ('N' if x != 'Y' else 'Y'))
```

```python
# 비고에 휴진 안내 + 점심시간 추가
df3_new['점심시간_평일'].fillna('없음', inplace = True)
df3_new['점심시간_토요일'].fillna('없음', inplace = True)
df3_new['점심시간'] = '[점심 시간]' + '\n' + '평일 : ' + df3_new['점심시간_평일'] + '\n토요일 : ' + df3_new['점심시간_토요일']
df3_new['비고'] = df3_new['휴진 안내'] + '\n\n' + df3_new['점심시간']
df3_new
```

```python
# 진료 시간 추출해서 날짜시간 형식으로 변환해줌
def convert_time(x, convert_type = 'start'):
    x = str(x)

    hour = x[:-4]
    min = x[-4:-2]
    if convert_type == 'start':
        text = str(hour) + ':' + str(min) + ':' + '00'
    else:
        text = str(hour) + ':' + str(min) + ':' + '59'

    return text


df3_new['일요일 진료시작시간'] = df3_new['진료시작시간_일'].apply(lambda x : convert_time(x, 'start'))
df3_new['일요일 진료종료시간'] = df3_new['진료종료시간_일'].apply(lambda x : convert_time(x, 'finish'))
df3_new['월요일 진료시작시간'] = df3_new['진료시작시간_월'].apply(lambda x : convert_time(x, 'start'))
df3_new['월요일 진료종료시간'] = df3_new['진료종료시간_월'].apply(lambda x : convert_time(x, 'finish'))
df3_new['화요일 진료시작시간'] = df3_new['진료시작시간_화'].apply(lambda x : convert_time(x, 'start'))
df3_new['화요일 진료종료시간'] = df3_new['진료종료시간_화'].apply(lambda x : convert_time(x, 'finish'))
df3_new['수요일 진료시작시간'] = df3_new['진료시작시간_수'].apply(lambda x : convert_time(x, 'start'))
df3_new['수요일 진료종료시간'] = df3_new['진료종료시간_수'].apply(lambda x : convert_time(x, 'finish'))
df3_new['목요일 진료시작시간'] = df3_new['진료시작시간_목'].apply(lambda x : convert_time(x, 'start'))
df3_new['목요일 진료종료시간'] = df3_new['진료종료시간_목'].apply(lambda x : convert_time(x, 'finish'))
df3_new['금요일 진료시작시간'] = df3_new['진료시작시간_금'].apply(lambda x : convert_time(x, 'start'))
df3_new['금요일 진료종료시간'] = df3_new['진료종료시간_금'].apply(lambda x : convert_time(x, 'finish'))
df3_new['토요일 진료시작시간'] = df3_new['진료시작시간_토'].apply(lambda x : convert_time(x, 'start'))
df3_new['토요일 진료종료시간'] = df3_new['진료종료시간_토'].apply(lambda x : convert_time(x, 'finish'))

df3_new['진료시작시간_일']
```

```python
# 진료시간 모두 null 값인 행 삭제
from ast import Continue
process_type = ['진료시작시간_일', '진료종료시간_일', '진료시작시간_월', '진료종료시간_월',
                '진료시작시간_화', '진료종료시간_화','진료시작시간_수', '진료종료시간_수',
                '진료시작시간_목', '진료종료시간_목','진료시작시간_금', '진료종료시간_금',
                '진료시작시간_토', '진료종료시간_토']

import math
for i in tqdm(range(len(df3_new))):
    count = 0
    for column in process_type:
        try:
            if math.isnan(df3_new.loc[i, column]):
                count += 1
        except:
            continue
    
    df3_new.loc[i, 'count_null'] = count
```

```python
df3_final = df3_new[df3_new['count_null'] < 14].reset_index(drop = True)
df3_final
```

```python
# 비고에 넣은 컬럼값 필요 없으니 삭제
df3_final.columns
drop_columns = ['일요일 휴진안내', '공휴일 휴진안내', '점심시간_평일', '점심시간_토요일'] + process_type + ['Unnamed: 30', 'Unnamed: 31']
df3_new = df3_final.drop(drop_columns, axis = 1).reset_index(drop = True)
df3_new.shape
```
