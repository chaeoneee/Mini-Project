## 1) matchType
+ matchType 별로 나눠서 모델 만들기
- solo 플레이어의 결과를 구할 때는 groupid, matchid, id, teamkills,revives, DBNOs 컬럼 불필요

```python
# 총 경기의 수(고유값의 수)
match_num = train.loc[:,"matchId"].nunique()
print(match_num)
```

```python
# matchType 별로 나눠서 예측
match_count = train['matchType'].value_counts().to_frame().reset_index()
match_count.columns = ["Type","Count"]
match_count
```

```python
# matchType별 경기 참여자 시각화
plt.figure(figsize=(10,6))
ticks = match_count.Type.values
ax = sns.barplot(x ="Type", y = "Count", data = match_count)
plt.ticklabel_format(axis ='y', useOffset = False, style = 'plain')
ax.set_xticklabels(ticks, rotation = 60, fontsize = 10)
ax.set_title("Match types")
plt.show()
```

```python
types = train["matchType"].value_counts().to_frame()
squads = types.loc[["squad-fpp","squad","normal-squad-fpp","normal-squad"],"matchType"].sum()
duos = types.loc[["duo-fpp","duo","normal-duo-fpp","normal-duo"],"matchType"].sum()
solo = types.loc[["solo-fpp","solo","normal-solo-fpp","normal-solo"],"matchType"].sum()
mt = pd.DataFrame([squads, duos, solo], index=["squad","duo","solo"], columns =["count"])
mt
```

```python
labels = ['squad', 'duo', 'solo'] ## 라벨
frequency = pd.Series.tolist(mt['count'])

fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성
 
ax.pie(frequency, ## 파이차트 출력
       labels=labels, ## 라벨 출력
       startangle=90, ## 시작점을 90도(degree)로 지정
       counterclock=False, ## 시계 방향으로 그린다.
       autopct=lambda p : '{:.2f}%'.format(p) ## 퍼센티지 출력
       )
 
plt.show()
```

```python
train.matchType.replace(['squad-fpp','squad','normal-squad-fpp','normal-squad'],'squad',inplace=True)
train.matchType.replace(['duo-fpp','normal-duo-fpp','normal-duo'],'duo',inplace=True)
train.matchType.replace(['solo-fpp','normal-solo-fpp','normal-solo'],'solo',inplace=True)
train.matchType.replace(['crashfpp','flaretpp','flarefpp','crashtpp'],'others',inplace=True)
```

## 2) movedistance(ride+swim+walk)
+ 평균 이동거리, 이동거리 범위 나누기
+ 이동거리와 kill 수의 관계
+ 이동거리와 등수의 관계

```python
train = pd.read_csv('train_V2.csv')
```

```python
# move distance
move_distance = train[['rideDistance','swimDistance','walkDistance']]
move_distance
```

```python
move_distance = move_distance.sum(axis = 1).to_frame()
move_distance.columns = ['moveDistance']
```

```python
move_distance = pd.concat([move_distance, train['kills']],axis = 1)
move_distance
```

```python
# 움직인 거리가 많으면 kills수도 많을 것이라고 생각했지만, 움직인 거리와 kills수는 관계 없음.
sns.set_style('whitegrid')
sns.set(rc = {'figure.figsize':(15,8)})
sns.scatterplot(data=move_distance, x='moveDistance', y='kills')#.set(xlim=(1000, 2000), ylim=(0, 40))
```

```python
move_distance['moveDistance'].max()  # 제일 많이 이동한 사람
```

```python
move_distance['moveDistance'].min()  # 제일 적게 이동한 사람(이상치로 제거해야 할듯)
```

```python
move_distance['moveDistance'].median() # 이동거리 중앙값
```

```python
move_distance['moveDistance'].mean() # 이동거리 평균
```

```python
pd.options.display.float_format = '{:.2f}'.format
move_df = move_distance['moveDistance'].describe().to_frame()
move_df
```

```python
move_win = pd.concat([move_distance, train['winPlacePerc']],axis = 1)
move_win
```

```python
# moveDistance가 클수록 승률이 높아질 것이다.

sns.jointplot(x="winPlacePerc", y="moveDistance",  data=move_win, height=10, ratio=3, color="lime")
plt.show()
```

### SCALING(성능 UP)
1. StandardScaler()
    + 모든 피처들을 평균이 0, 분산이 1인 정규분포를 갖도록 함
    + 데이터 내에 이상치가 있으면, 데이터의 평균과 분산에 크게 영향을 줌
2. MinMaxScaler()
    + 모든 피처들이 0과 1사이의 데이터 값을 갖도록 함(피처별로 최솟값은 0이 되고, 최댓값은 1)
    + 이상치가 존재한다면, 이상치가 극값이 되어 데이터가 아주 좁은 범위에 분포하게 됨
