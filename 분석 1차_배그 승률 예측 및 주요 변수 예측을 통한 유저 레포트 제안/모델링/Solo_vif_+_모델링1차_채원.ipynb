# Scaling

```python
# 데이터 프레임 전체에 scaling 적용
# from sklearn.preprocessing import StandardScaler #이상치가 많을 때 사용 지양
# from sklearn.preprocessing import MinMaxScaler
# from sklearn.preprocessing import Normalizer # 정규분포 아니거나 표준 편차 작을 때 good
# from sklearn.preprocessing import RobustScaler #  이상치 영향 적음
```

```python
#스케일링 : 변수간의 단위 보정
from sklearn.preprocessing import RobustScaler
scaler = ['boosts', 'damageDealt', 'headshotKills', 'heals', 'killPlace', 'kills', 'killStreaks', 'longestKill', 'matchDuration', 'numGroups', 'roadKills',
       'vehicleDestroys', 'walkDistance', 'weaponsAcquired','totalDistance', 'cnt_per_matchid', 'total_heals', 'items', 'teamwork',
       'headshots_over_kills', 'killPlace_over_maxPlace', 'workDistance_over_kills', 'kills_over_match_cnt']
min = RobustScaler()
min.fit(df[scaler])
df[scaler] = min.transform(df[scaler])
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('winPlacePerc', axis = 1)
y = df['winPlacePerc']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
X_train
```

## VIF_파생변수 넣기 전 
### 변수 선정 
['kills', 'killStreaks', 'boosts', 'walkDistance', 'killPlace',       'headshotKills', 'longestKill', 'weaponsAcquired', 'heals',       'matchDuration', 'numGroups', 'roadKills', 'vehicleDestroys']

```python
# 파생변수 넣기 전 VIF 

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_drop = ['killPlace_over_maxPlace','headshots_over_kills','workDistance_over_kills','totalDistance','teamwork','cnt_per_matchid','kills_over_match_cnt','total_heals','items','damageDealt']
solo_vif = X_train.drop(vif_drop, axis = 1)
solo_vif_t = X_test.drop(vif_drop, axis = 1)

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(solo_vif.values, i) for i in range(solo_vif.shape[1])]
vif['features'] = solo_vif.columns
vif = vif.sort_values('VIF Factor', ascending=False).reset_index(drop=True)
vif
```

```python
vif['features'].values
```

```python
X_train_1 = solo_vif
X_test_1 = solo_vif_t
```

```python
# solo_vif.corr().style.background_gradient().set_properties(**{'font-size': '11pt'})
solo_vif = solo_vif.corr()
# 그림 사이즈 지정
fig, ax = plt.subplots( figsize=(15,15) )

# 삼각형 마스크를 만든다(위 쪽 삼각형에 True, 아래 삼각형에 False)
mask = np.zeros_like(solo_vif, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# 히트맵을 그린다
sns.heatmap(solo_vif, 
            cmap = 'RdYlBu_r', 
            annot = True,   # 실제 값을 표시한다
            mask=mask,      # 표시하지 않을 마스크 부분을 지정한다
            linewidths=.5,  # 경계면 실선으로 구분하기
            cbar_kws={"shrink": .5},# 컬러바 크기 절반으로 줄이기
            vmin = -1,vmax = 1,
            annot_kws={"size": 15}# 컬러바 범위 -1 ~ 1
           )  
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model_lr = LinearRegression(normalize=True, n_jobs=8)
model_lr = model_lr.fit(X_train_2, y_train)

print("Linear Train Score:", model_lr.score(X_train_2, y_train))
print("Linear Test Score:", model_lr.score(X_test_2, y_test))

pred_lr = model_lr.predict(X_test_2)  # test데이터의 예측값
```

```python
from sklearn.ensemble import RandomForestRegressor

size = 0.20
seed = 42
model_rf = RandomForestRegressor(n_estimators=10)

model_rf = model_rf.fit(X_train_2,y_train)

print("Random Forest Train Score:", model_rf.score(X_train_2, y_train))
print("Random Forest Test Score:", model_rf.score(X_test_2, y_test))

pred_rf = model_rf.predict(X_test_2)
```

```python
import lightgbm as lgb
train_lgb = lgb.Dataset(X_train_2, label = y_train) 
test_lgb = lgb.Dataset(X_test_2, label = y_test) 

params = {'learning_rate': 0.01, 
          'max_depth': 16, 
          'boosting': 'gbdt', 
          'objective': 'regression', 
          'metric': 'mse', 
          'is_training_metric': True, 
          'num_leaves': 144, 
          'feature_fraction': 0.9, 
          'bagging_fraction': 0.7, 
          'bagging_freq': 5, 
          'seed':2020}

model_lgb = lgb.train(params, train_lgb, 1000, test_lgb, verbose_eval=100, early_stopping_rounds=500)
predict_lgb_train = model_lgb.predict(X_train_2)
predict_lgb_test = model_lgb.predict(X_test_2)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error(y_test, predict_lgb_test)
r2 = r2_score(y_test, predict_lgb_test)
print('lgb result')
print('mse : ',mse)
print('r2 : ',r2)
```

```python
# 앙상블(스태킹)
from sklearn.ensemble import StackingRegressor
clf = StackingRegressor(estimators=[('lr',model_lr),('rf',model_rf)],final_estimator = RandomForestRegressor())
clf.fit(X_train_2, y_train)
pred_st = clf.predict(X_test_2)
```
