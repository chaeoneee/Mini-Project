```python
re_df['REPLY']=re_df['REPLY'].astype('str')
```

```python
arr = []
for reply in re_df['REPLY']:
    cache = kiwi.tokenize(reply)
    for token in cache:
        if token[1].startswith('NN'):
            arr.append(token[0])
```

```python
#토픽모델링

arr = []
for reply in re_df['REPLY']:
    cache = kiwi.tokenize(reply)
    brr = []
    for token in cache:
        if token[1] == 'NNG':
            brr.append(token[0])
        elif token[1] == 'NNP':
            brr.append(token[0])
        elif token[1] == 'VV':
            brr.append(token[0]+'다')
        elif token[1] == 'VA':
            brr.append(token[0]+'다')
#        elif token[1] == 'SL':
#            brr.append(token[0]) #이 부분이 영어입니다.
        elif token[1] == 'XR':
            brr.append(token[0])
    arr.append(brr)
            
print(arr)
```

```python
drr = []
for i in arr:
    drr.append(len(i))
    
print(max(drr))
```

```python
tokenized_arr = []
for words in arr:
    words_list = []
    if len(set(words)) > 2:
        words_str = ' '.join(words)
        tokenized_arr.append(words_str)

tokenized_arr
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
#LDA 는 Count기반의 Vectorizer만 적용 
count_vectorizer = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1,2))
    # 2개의 문서 미만으로 등장하는 단어는 제외, 전체의 10% 이상으로 자주 등장하는 단어는 제외
    # bigram도 포함

feat_vect = count_vectorizer.fit_transform(tokenized_arr)
print('CountVectorizer Shape:', feat_vect.shape)
```

```python
lda = LatentDirichletAllocation(n_components=5)  # 토픽 수는 6개로 설정
lda.fit(feat_vect)
```

```python
def display_topics(model, feature_names, num_top_words):
    for topic_index, topic in enumerate(model.components_):
        print('Topic #', topic_index)

        # components_ array에서 가장 값이 큰 순으로 정렬했을 때, 그 값의 array index를 반환. 
        topic_word_indexes = topic.argsort()[::-1]
        top_indexes=topic_word_indexes[:num_top_words]
        
        # top_indexes대상인 index별로 feature_names에 해당하는 word feature 추출 후 join으로 concat
        feature_concat = ' '.join([feature_names[i] for i in top_indexes])                
        print(feature_concat)

# CountVectorizer객체내의 전체 word들의 명칭을 get_features_names( )를 통해 추출
feature_names = count_vectorizer.get_feature_names_out()

# Topic별 가장 연관도가 높은 word를 10개만 추출
display_topics(lda, feature_names, 20)
```
