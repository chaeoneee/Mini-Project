```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
#kiwi.load_user_dictionary('user_dictionary.txt')
kiwi.prepare()
```

```python
re_df = pd.read_csv('댓글크롤링_피식대학_피식쇼_르세라핌.csv')
```

```python
re_df=re_df[['ID','REPLY','LIKE','LIKE2']]
re_df
```

```python
re_df.info()
```

```python
re_df['REPLY']=re_df['REPLY'].astype('str')
```

```python
arr = []
for reply in re_df['REPLY']:
    cache = kiwi.tokenize(reply)
    for token in cache:
        if token[1].startswith('NN'):
            arr.append(token[0])
```

```python
from collections import Counter
Counter(arr).most_common()
```

```python
#토픽모델링

arr = []
for reply in re_df['REPLY']:
    cache = kiwi.tokenize(reply)
    brr = []
    for token in cache:
        if token[1] == 'NNG':
            brr.append(token[0])
        elif token[1] == 'NNP':
            brr.append(token[0])
        elif token[1] == 'VV':
            brr.append(token[0]+'다')
        elif token[1] == 'VA':
            brr.append(token[0]+'다')
#        elif token[1] == 'SL':
#            brr.append(token[0]) #이 부분이 영어입니다.
        elif token[1] == 'XR':
            brr.append(token[0])
    arr.append(brr)
            
print(arr)
```

```python
drr = []
for i in arr:
    drr.append(len(i))
    
print(max(drr))
```

```python
tokenized_arr = []
for words in arr:
    words_list = []
    if len(set(words)) > 2:
        words_str = ' '.join(words)
        tokenized_arr.append(words_str)

tokenized_arr
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
#LDA 는 Count기반의 Vectorizer만 적용 
count_vectorizer = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1,2))
    # 2개의 문서 미만으로 등장하는 단어는 제외, 전체의 10% 이상으로 자주 등장하는 단어는 제외
    # bigram도 포함

feat_vect = count_vectorizer.fit_transform(tokenized_arr)
print('CountVectorizer Shape:', feat_vect.shape)
```

```python
lda = LatentDirichletAllocation(n_components=6)  # 토픽 수는 6개로 설정
lda.fit(feat_vect)
```

```python
LatentDirichletAllocation(batch_size=128, doc_topic_prior=None,
             evaluate_every=-1, learning_decay=0.7,
             learning_method='batch', learning_offset=10.0,
             max_doc_update_iter=100, max_iter=10, mean_change_tol=0.001,
             n_components=6, n_jobs=None, perp_tol=0.1,
             random_state=None, topic_word_prior=None,
             total_samples=1000000.0, verbose=0)
```

```python
def display_topics(model, feature_names, num_top_words):
    for topic_index, topic in enumerate(model.components_):
        print('Topic #', topic_index)

        # components_ array에서 가장 값이 큰 순으로 정렬했을 때, 그 값의 array index를 반환. 
        topic_word_indexes = topic.argsort()[::-1]
        top_indexes=topic_word_indexes[:num_top_words]
        
        # top_indexes대상인 index별로 feature_names에 해당하는 word feature 추출 후 join으로 concat
        feature_concat = ' '.join([feature_names[i] for i in top_indexes])                
        print(feature_concat)

# CountVectorizer객체내의 전체 word들의 명칭을 get_features_names( )를 통해 추출
feature_names = count_vectorizer.get_feature_names_out()

# Topic별 가장 연관도가 높은 word를 10개만 추출
display_topics(lda, feature_names, 20)
```

```python
from gensim.models.ldamodel import LdaModel
from gensim.models.callbacks import CoherenceMetric
from gensim import corpora
from gensim.models.callbacks import PerplexityMetric

import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
arr
```

```python
dictionary = corpora.Dictionary(arr)
```

```python
corpus = [dictionary.doc2bow(text) for text in arr]
```

```python
num_topics = 5
chunksize = 2000
passes = 20
iterations = 400
eval_every = None

temp = dictionary[0]
id2word = dictionary.id2token

model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)
```

```python
top_topics = model.top_topics(corpus) #, num_words=20)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

from pprint import pprint
pprint(top_topics)
```
