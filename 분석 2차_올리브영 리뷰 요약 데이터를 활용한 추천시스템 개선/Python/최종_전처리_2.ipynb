# **crawl_done.pickle**

```python
with open("crawl_done.pickle", "rb") as g:
    review = pickle.load(g) #불러오기
```

```python
# 크롤링한 리뷰 중 사용 가능한 리뷰 : 상품명이 null값이 아닌 경우 & 중복값 제외
df = review[review['상품'].isna() == False].reset_index(drop=True)
df = df[df.duplicated()==False].reset_index(drop=True)
df.shape
```

```python
df.head(5)
```
